### Теория
Задача заключается в разделении наблюдений на исходные сигналы, без знания о том, как именно наблюдения были составлены из набора исходных сигналов.

Требуется принять во внимание следующие предположения:
- Исходные данные должны быть покомпонентно неотрицательными векторами
- В каждом из наблюдений должен превалировать какой-то из исходных сигналов
- Сумма элементов в каждом ряде смешивающей матрицы должна быть равна единице
- Размерность вектора наблюдения должна быть не мнеьше размерности исходного вектора и должна быть равна столбцовому рангу смешивающей матрицы

По сути, если считать $s$ исходным сигналом, а $x$ считать наблюдением, мы решаем следующую задачу:
$$
x = As,
$$

Задача может быть не выпуклой, но мы можем ее приблизить, чтобы пользоваться методами выпуклой оптимизации. Собственно это мы и сделаем.

```python
import numpy as np
import scipy.stats as sps
import cvxpy as cvx
from glob import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

Ниже моя реализация CAMNS.

P.S. Я пытался сделать оптимизацию с удалением нулевых векторов, но она тут не очень-то и нужна, а еще меня убило об размерности и я забил

```python
def is_ext_pt(C, d, alpha, tol):
    L, D = C.shape
    T = C[np.all(np.abs(np.expand_dims(C.dot(alpha) + d, axis=1)) < tol, axis=1), :]
    return T.shape[0] != 0 or np.linalg.matrix_rank(T, tol=tol) == D

def CAMNS(X, N, tol_lp=1e-3, tol_ext=1e-8, tol_zeros=1e-6):
  # Getting shape of the input signal.
  M, L = X.shape
  # Since I don't want to make regular Python users cry, matrix is passed in a
  # pythonic way, so we need to transpose it for the algo to work.
  X = X.T
  d =  np.mean(X, axis=1)
  C, V, S = np.linalg.svd(X - np.expand_dims(d, axis=1), full_matrices=False)
  # Drop the unneeded one.
  C = C[:, :N-1]

  l = 0

  # ----
  # step 1
  l = 0
  hS = np.zeros(shape=(0, L))
  # Initalize B with identity matrix.
  B = np.diag(np.ones(L))
  # ----
  # Repeating untill we get exactly at least N signals.
  while l < N:
    # step 2
    w = sps.norm.rvs(size=L)
    r = B.dot(w)  
    # ----
    # step 3: Solving convex optimizaton problems
    alpha_1 = cvx.Variable(C.shape[1])
    goal_1 = cvx.Minimize(r.T @ (C @ alpha_1 + d))
    problem_1 = cvx.Problem(goal_1, [C @ alpha_1 + d >= 0])
    p = problem_1.solve(solver='ECOS_BB')
    a_1 = alpha_1.value

    alpha_2 = cvx.Variable(C.shape[1])
    goal_2 = cvx.Maximize(r.T @ (C @ alpha_2 + d))
    problem_2 = cvx.Problem(goal_2, [C @ alpha_2 + d >= 0])
    q = problem_2.solve(solver='ECOS_BB')
    a_2 = alpha_2.value
    # ----
    # step 4
    if l == 0:
      if is_ext_pt(C, d, a_1, tol_ext):
        hS = np.append(hS, [(C @ a_1 + d).squeeze()], axis=0)
      if is_ext_pt(C, d, a_2, tol_ext):
        hS = np.append(hS, [(C @ a_2 + d).squeeze()], axis=0)
    else:
      if p / (np.linalg.norm(r) * np.linalg.norm(C @ a_1 + d)) >= tol_lp:
        if is_ext_pt(C, d, a_1, tol_ext):
          hS = np.append(hS, [(C @ a_1 + d).squeeze()], axis=0)
      if q / (np.linalg.norm(r) * np.linalg.norm(C @ a_2 + d)) >= tol_lp:
        if is_ext_pt(C, d, a_2, tol_ext):
          hS = np.append(hS, [(C @ a_2 + d).squeeze()], axis=0)
    # ----
    # step 5: Updating the number of vectors
    l = hS.shape[0]
    # ----
    #step 6: Applying QR decomposition
    if l > 0:
      Q, _ = np.linalg.qr(hS.T)
      B = np.diag(np.ones(L)) - Q @ Q.T
    # ----
    print(l)
  return hS
```

Загрузим картинки, перенесем их в черно-белую палитру и перемешаем сигналы

```python
paths = glob('img/*')
N_IMAGES = len(paths)

fig, ax = plt.subplots(2, N_IMAGES, figsize=(10, 10))
images = []
for i in range(N_IMAGES):
    cur_img = np.array(Image.open(paths[i]).convert('L'))
    ax[0][i].imshow(cur_img, cmap='gray')
    images.append(cur_img.flatten().astype(np.float64) / 255.0)

images = np.array(images)
random_matrix = np.random.rand(N_IMAGES, N_IMAGES)
mixing_matrix = np.divide(random_matrix, np.sum(random_matrix, axis=1, keepdims=True))

print(N_IMAGES)

mixed_channels = mixing_matrix.dot(images)
for i in range(N_IMAGES):
    ax[1][i].imshow(mixed_channels[i].reshape(64, 64), cmap='gray')
plt.show()
```

А теперь восстановим сигналы обратно. Восстановление часто зависит от смешения, поэтому можно поперезапускать эти две ячейки, если оно восстановилось плохо

```python
fig, ax = plt.subplots(1, N_IMAGES, figsize=(10, 10))
unmixed_channels = CAMNS(mixed_channels, N_IMAGES)
for i in range(N_IMAGES):
    ax[i].imshow(unmixed_channels[i].reshape(64, 64),cmap='gray')
plt.show()
```

Сигналы восстановились почти в точности, а значит алгосик работает. Можно посмотреть на расстояние до изначальных картинок.

```python
for i in range(N_IMAGES):
    print(np.linalg.norm(images[i] - unmixed_channels[i]) / (np.linalg.norm(images[i]) * np.linalg.norm(unmixed_channels[i])))
```

Как видно, совпадение достаточно большое, что еще раз подтверждает корректность работы алгоритма.
